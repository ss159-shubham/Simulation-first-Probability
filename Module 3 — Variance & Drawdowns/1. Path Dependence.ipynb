# Module 3.1 Path Dependence: 
Theme: *Order Matters More then Outcome*

### Background assumpthions
- The set of outcomes is fixed.
- The distribution of wins and losses does not change
- Expectancy is held constant across simulations

These assumptions are treated as given and are not re-examined in this module.

### Purpose of this simulation
To demonstrate that sequencing alone, without changing outcomes or probabilities,    
can produce substantially different equity trajectories and lived experiences.

### Common intuitive beliefs (to be examined)
- Equity paths with the same final outcome should feel similar
- Early performance is informative about long-term success
- Adverse runs reflect inferior conditions rather than sequencing

### What this module isolates
- Outcome order as the only variable
- Path-level behavior instead of summary statistics

### Objective
> To observe how identical outcomes, arranged differently in time,   
create materially different trajectories and drawdowns.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Defining fixed outcome set

```python
n_trades = 500
# Example: +1 for win, -1 for loss
# The exact probabilities are irrelevant â€” they are frozen
wins = 275
losses = n_trades - wins

outcomes = np.array([1]*wins + [-1]*losses)
```

# Single path generator

```python
def generate_equity_path(outcomes):
    shuffled = np.random.permutation(outcomes)
    equity = np.cumsum(shuffled)
    return equity

```

# Generate Multiple Paths

```python
n_paths = 20
paths = [generate_equity_path(outcomes) for _ in range(n_paths)]
```

```python
plt.figure(figsize=(10, 6))

for path in paths:
    plt.plot(path, alpha=0.7)

plt.axhline(0 , ls = "--")
plt.title("Same Outcomes, Different Orders â†’ Different Lives")
plt.xlabel("Trade Number")
plt.ylabel("Cumulative Outcome")
plt.show()
```

```python
final_equities = [path[-1] for path in paths]

plt.figure(figsize=(6, 4))
plt.hist(final_equities, bins=10)
plt.title("Final Outcomes Across Paths")
plt.xlabel("Final Equity")
plt.ylabel("Frequency")
plt.show()
```

```python
def max_drawdown(equity):

    peak = equity[0]
    max_dd = 0

    for x in equity:
        if x > peak:
            peak = x
        
        dd = peak - x
        max_dd = max(max_dd, dd)
    
    return max_dd
        

drawdowns = [max_drawdown(path) for path in paths]

plt.figure()
plt.hist(drawdowns)
plt.xlabel('drawdown length')
plt.ylabel('n_paths')
plt.title('diffrerent drawdown life with same outcome')
plt.show()
```

# Conclusion

- Some paths felt broken early
- Some paths felt effortless
- Some paths tested patience before any reward

And yet:
- Outcomes were identical
- No rule changed
- No edge changed

#### Insight
> The market did not treat agents differently.     
> **The order did**
