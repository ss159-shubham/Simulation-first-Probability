# Module 2.7 — Mathematical Expectancy vs Lived Experience
### Pre-Simulation beliefs:
1. If a system has positive expectancy, it should feel like progress most of the time.
2. New equity highs should occur regularly.
3. Emotional discomfort should mostly coincide with genuine system underperformance.
4. Time spent “going nowhere” should be relatively brief in a winning system.

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.55
n_trades = 1000
```

```python
seq = np.where(
    np.random.rand(n_trades) < p,
    1, -1
)

equity = np.cumsum(seq)
equity_high = np.maximum.accumulate(equity)
drawdown = equity - equity_high

# time spent below high
below_high = (drawdown < 0).astype(int)
percent_time_below = below_high.mean() * 100

print(f"Time spent below equity highs: {percent_time_below:.2f}%")
```

### A winning system spends most of its life not at new highs.

```python
plt.figure(figsize = (12, 5))
plt.plot(equity, label = 'Equity Curve', alpha = 0.9)
plt.plot(equity_high, label = 'Equity High', alpha = 0.7, linestyle = "--")
plt.legend()
plt.xlabel('Trades')
plt.ylabel('P&L')
plt.title('Equity Curve vs Running High')
plt.tight_layout()
plt.show()
```

## Drawdown curve

```python
plt.figure(figsize=(12, 4))
plt.plot(drawdown, color='red', alpha=0.8)
plt.axhline(0, linestyle='--')
plt.xlabel('Trades')
plt.ylabel('Drawdown')
plt.title('Drawdown Curve (Psychological Experience)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(drawdown, bins=50)
plt.xlabel('Drawdown')
plt.ylabel('Frequency')
plt.title('Distribution of Drawdown Depths')
plt.tight_layout()
plt.show()
```

# Observation
### This simulation shows a sharp divergence between mathematical expectancy and lived experience. While the equity curve trends upward over the long run, the system spends the majority of its time below prior equity highs, embedded in drawdowns or flat periods. These periods dominate the trader’s day-to-day emotional experience despite being fully consistent with a functioning positive-expectancy process. The reward of expectancy arrives intermittently through new equity highs, while the psychological cost is paid continuously through stagnation and drawdown. Emotional discomfort, therefore, is not evidence of system failure, but a structural feature of participating in probabilistic systems.
