# Module 2.1 — What “Positive Expectancy” Actually Means (and Doesn’t)

### Pre-Simulation beliefs

1. A system with positive expectancy should:
   - (e.g.) mostly go up over time
   - (e.g.) feel reliable after enough trades    

2. If expectancy is positive, I expect:
   - Drawdowns to be limited
   - Losing streaks to be relatively rare
   - Confidence to increase with time       

3. If I experienced a long flat or losing period, my instinct would be to:
   - Question the system
   - Assume something changed
   - Feel that expectancy is not “working”       


```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.55
n_trials = 1000
```

## Single Long Run (The Human Experience)

```python
#np.random.seed(42)

outcomes = np.where(np.random.rand(n_trials) < p, 1, -1)
equity = outcomes.cumsum()

plt.figure()
plt.plot(equity)
plt.xlabel('no. of trials')
plt.ylabel('equity')
plt.title('Equity curve')
plt.show()
```

## Many Independent Runs (The Population)

```python
def run_simulation(p, n_trials):
    outcomes = np.where(np.random.rand(n_trials) < p, 1, -1)
    return outcomes.cumsum()

n_runs = 500
final_equities = []
all_paths = []


for _ in range(n_runs):
    path = run_simulation(p, n_trials)
    all_paths.append(path)
    final_equities.append(path[-1])

final_equities = np.array(final_equities)

plt.figure()
plt.hist(final_equities)
plt.title('distribution of final equities on all paths')
plt.xlabel('no. of Paths')
plt.ylabel('final equity')
plt.show()
```

```python
max_equity = max(final_equities)
min_equity = min(final_equities)
av_equity = final_equities.mean()

print(f"max final equity = {max_equity}\nmin final equity = {min_equity}\nAverage final equity = {av_equity}")
print('Almost all agents had positive outcome.')
```

# Observation

### This simulation demonstrates that randomness itself is not inherently harmful. In a controlled environment—characterized by bounded outcomes, additive dynamics, and no survival constraints—randomness produces broadly favorable results when the underlying probability is biased in a positive direction. Increasing this bias does not eliminate variability or guarantee outcomes for any single agent, but it shifts the entire distribution of results toward favorable regions.

### The simulation also highlights that pure randomness is not exploitable, as it contains no structure or bias to act upon. Exploitability emerges only when randomness is constrained or shaped by structural rules. In such cases, randomness becomes directional rather than neutral—maneuverable rather than adversarial. Adverse outcomes, therefore, are not an inevitable consequence of randomness itself, but arise from how randomness is structured, amplified, or compounded within a system.
