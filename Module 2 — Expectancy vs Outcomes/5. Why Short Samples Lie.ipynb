# Module 2.5 — Why Short Samples Lie
### Pre-Simulation beiefs
1. Early results should be somewhat indicative
    - If a system is good, it should “start showing it” fairly quickly.
    - Large early losses or gains must mean something about quality.
2. Extremes imply information
    - Strong early profits suggest skill or a strong edge.
    - Strong early losses suggest flaws, bad execution, or a broken model.
3. Variance should calm down quickly
    - After a modest number of trades, results should feel more stable.
    - Wide swings early on feel abnormal or concerning.
4. Emotional reactions are proportional to information
    - Confidence after winning streaks feels justified.
    - Doubt after losing streaks feels diagnostic.

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.55
n_trades = 1000
n_agents = 5000
```

```python
seqs = np.where(
    np.random.rand(n_agents, n_trades) < p,
    1, 
    -1
)
paths = np.cumsum(seqs, axis = 1)
N = [20, 500 , 1000]

final_eq = {}

for i in N:
    thres_path = paths[:, :i]
    final_equity = thres_path[:, -1]
    final_eq[i] = final_equity
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (15,5))
ax[0].hist(final_eq[20])
ax[0].set_title('Short horizon (N = 20)')
ax[0].set_xlabel('PnL')
ax[0].set_ylabel('No. of agents')

ax[1].hist(final_eq[500])
ax[1].set_title('Long horizon (N = 500)')
ax[1].set_xlabel('PnL')

fig.tight_layout()
plt.show()
```

# Observation
### As sample size increases, outcome distributions shift toward the positive expectancy, and relative dispersion contracts. However, absolute outcome variability continues to grow with scale, allowing for increasingly extreme realizations on both the positive and negative sides. This means that while the probability of loss declines with time, the magnitude of worst-case outcomes does not vanish. Short samples therefore lie by exaggerating confidence and despair, while long samples clarify expectancy without eliminating risk.
### Short samples produce the strongest emotions and the weakest information.
