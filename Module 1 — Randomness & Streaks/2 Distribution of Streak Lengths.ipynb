# Project 1.2 — Streak Length Distributions

### Belief (before simulation):
    - Long streaks should be rare
    - A 10-loss streak feels almost impossible
    - Streaks imply something changed


```python
import numpy as np
import matplotlib.pyplot as plt

n_flips = 1000
outcomes = np.random.choice([1, -1], size=n_flips)
```

```python
streaks = []
current_streak = 1

for i in range(1, len(outcomes)):
    if outcomes[i] == outcomes[i - 1]:
        current_streak += 1
    else:
        streaks.append(current_streak)
        current_streak = 1

streaks.append(current_streak)

print("Total streaks:", len(streaks))
print("Max streak:", max(streaks))
```

```python
plt.figure()
plt.hist(streaks, bins = range(1, max(streaks) + 1), align = 'left')
plt.xlabel("Streak length")
plt.ylabel("Frequency")
plt.title("Streak Length Distribution (Fair Coin)")
plt.show()
```

```python
for k in [5, 8, 10, 12, 15]:
    print(f"Streaks ≥ {k}:", sum(s >= k for s in streaks))
```

# observation
### In a purely random 50–50 process, long streaks of identical outcomes occur frequently over large samples. Although any single long streak feels unlikely, the cumulative probability across many trials makes such streaks inevitable. Emotional responses to these streaks — including doubt and quitting — are reactions to normal variance, not to changes in underlying probabilities.
