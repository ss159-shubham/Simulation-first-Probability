# Project 1.1 - Fair Coin ≠ Smooth Outcomes

### Belief (before simulation):
    - A fair coin should balance out
    - The cumulative result should not drift too much
    - Long ugly runs should be rare.
    

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed()

n_flips = 1000

# 1 = win , -1 = loss
outcomes = np.random.choice([1, -1], size = n_flips)

cumulative = np.cumsum(outcomes)

plt.figure()
plt.plot(cumulative)
plt.axhline(0)
plt.xlabel("Time (flips)")
plt.ylabel("Cumulative outcome")
plt.title("Fair Coin — 1000 Flips")
plt.show()
```

# Multiple independent paths

```python
n_paths = 20

plt.figure()

for _ in range(n_paths):
    outcomes = np.random.choice([1, -1], size=n_flips)
    cumulative = np.cumsum(outcomes)
    plt.plot(cumulative)

plt.axhline(0)
plt.xlabel("Time (flips)")
plt.ylabel("Cumulative outcome")
plt.title("20 Fair Coin Paths — Same Odds, Different Lives")
plt.show()
```

# observation
### A fair random process does not produce smooth or visually balanced paths. Individual outcome paths can drift strongly upward or downward for long periods without any underlying bias. Fairness exists only at the distribution level across many possible paths, not at the level of a single realized path.

### In other words, multiple agents can do the same job with the same probability of outcome and can have drastically different individual outcomes.

### This project answers only one question:
### *“Can randomness alone create long-term-looking inequality?”*

### The answer is yes.
