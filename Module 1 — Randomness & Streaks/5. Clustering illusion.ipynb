# Project 1.5 - Clustering illusion
### Beliefs(before simulation)
    - If outcomes cluster visually over time, there is likely a structural reason or change in the underlying process causing it.
    - Randomness should “mix”.
    - Independence should look scattered.
    - Repetition should be rare.
    - Clusters should require explanation.
    - "Pure randomness should look evenly spread, not clumped."

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
seq = np.random.choice([0, 1], size = n)

plt.figure()
plt.bar(range(n), seq, width=1.0)
plt.yticks([0,1])
plt.title("Clustering of Binary Outcomes Over Time")
plt.show()
```

```python
cum = np.cumsum(np.where(seq == 1, 1, -1))
plt.figure()
plt.plot(cum)
plt.axhline(0, linestyle='--')
plt.title("Cumulative Path of Random Outcomes")
plt.show()
```

# Outcome
### The clustering of outcomes over time does not imply causality or structural change. Even in fully random and memoryless processes, outcomes naturally form clusters that can appear logical or meaningful when viewed visually or cumulatively. These patterns arise from repetition within randomness, not from underlying reasons or regimes.
