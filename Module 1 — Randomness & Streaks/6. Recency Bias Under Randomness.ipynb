### belief (before simulation)
- *If the last several outcomes were the same (e.g., 5 losses in a row), the next outcome is more likely to reverse or continue.*


```python
import numpy as np

n = 1_000_000
seq = np.random.choice([0, 1], size = n)

condition_length = 5
next_outcomes = []

for i in range(condition_length, n):
    window = seq[i-condition_length : i]
    if np.all(window == window[0]):
        next_outcomes.append(seq[i])

next_outcomes = np.array(next_outcomes)
```

#### *the probability of outcome after the same previous outcomes*

```python
next_outcomes.mean()
```

# Observation 

### Conditioning on recent outcomes in a memoryless random process does not change the probability of the next outcome. Despite strong intuitive expectations of reversal or continuation after streaks, recent history contains no predictive information. The belief that it does reflects cognitive bias rather than statistical reality.
