# Project 1.4 â€” Short Samples vs Long Samples
### Belief (before simulation):
    - short samples --> noisy but "reasonable"
    - long samples --> smoother and safer

```python
import numpy as np
```

```python
def max_streak(seq):
    
    max_run = run = 1  # run --> length of current streak, max_run --> length of the longest streak 

    for i in range(1, len(seq)):
        if seq[i] == seq[i-1]:
            run += 1
            max_run = max(max_run, run)
        else:
            run = 1

    return max_run


def simulate(sample_length, n_paths = 500):
    max_streaks = []
    for _ in range(n_paths):
        seq = np.random.choice([-1,1], size = sample_length)
        max_streaks.append(max_streak(seq))

        return max_streaks


print('short:', simulate(20))
print('long: ',simulate(100)) 
```

# Observation
### Long streaks occur in both short and long random samples, with longer samples typically containing longer maximum streaks. However, in short samples, a single streak represents a large fraction of the total history and therefore dominates perception. In long samples, even more extreme streaks are present, but their contribution to the total sample is proportionally smaller and spread across time, making the overall experience appear calmer. As a result, extreme streaks are not prevented by short sample lengths, and shorter histories are not inherently safer or more representative.
