# Project 1.8 — Same Distribution, Different Lives
### belief (before simulation):
#### *“If multiple people play the same game, face the same odds for the same amount of time, their outcomes should look roughly similar.”*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_paths(n_paths, n_steps):
    paths = []
    seq = []
    for _ in range(n_paths):
        steps = np.random.choice([1, -1], size = n_steps)
        cum = np.cumsum(steps)
        paths.append(cum)
        seq.append(steps)
    return np.array(paths) , np.array(seq)
```

```python
n_agents = 1000
n_steps = 500

paths, _ = simulate_paths(n_paths=n_agents, n_steps=n_steps)

final_outcomes = paths[:, -1]
```

## The distribution of lives

```python
plt.figure()
plt.hist(final_outcomes, bins=40)
plt.xlabel("Final Outcome")
plt.ylabel("Number of Agents")
plt.title("Same Probabilities, Different Lives")
plt.show()
```

## comparing paths

```python
best = np.argmax(final_outcomes)
worst = np.argmin(final_outcomes)
median = np.argsort(final_outcomes)[len(final_outcomes)//2]

plt.figure()
plt.plot(paths[best], label="Best Life")
plt.plot(paths[median], label="Median Life")
plt.plot(paths[worst], label="Worst Life")
plt.legend()
plt.title("Identical Odds, Violently Different Lives")
plt.show()
```

# Observation

### When multiple agents play the same game with identical probabilities over the same time horizon, the distribution of final outcomes forms a bell-shaped curve. However, this aggregate symmetry masks extreme individual differences. Agents located near opposite tails of the distribution experience drastically different outcomes despite facing identical odds.

### These differences do not arise from skill, decision quality, or structural advantage, but purely from path dependence within the same probabilistic system. As a result, individual outcomes are poor indicators of underlying ability when viewed in isolation.
